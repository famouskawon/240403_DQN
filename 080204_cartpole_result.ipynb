```python
import pickle
import numpy as np # type: ignore
import os

loaded_reward_histories = []

# 저장된 파일 불러오기
index = 0
filename = f'080205_CartpoleResults/DQN_reward_history_{index}.pkl'
while os.path.exists(filename):

    with open(f'080205_CartpoleResults/DQN_reward_history_{index}.pkl', 'rb') as f:
        reward_history = pickle.load(f)
        loaded_reward_histories.append(reward_history)
    index += 1  # 파일이 이미 존재하면 인덱스 증가
    filename = f'080205_CartpoleResults/DQN_reward_history_{index}.pkl'  # 새로운 파일 이름 업데이트
DQN = np.mean(loaded_reward_histories,axis=0)  # 각 리스트는 임의의 보상 값 10개를 포함
```

```python
loaded_reward_histories = []

# 저장된 파일 불러오기
index = 0
filename = f'080205_CartpoleResults/Double_DQN_reward_history_{index}.pkl'
while os.path.exists(filename):

    with open(f'080205_CartpoleResults/Double_DQN_reward_history_{index}.pkl', 'rb') as f:
        reward_history = pickle.load(f)
        loaded_reward_histories.append(reward_history)
    index += 1  # 파일이 이미 존재하면 인덱스 증가
    filename = f'080205_CartpoleResults/Double_DQN_reward_history_{index}.pkl'  # 새로운 파일 이름 업데이트

Double_DQN = np.mean(loaded_reward_histories,axis=0)  # 각 리스트는 임의의 보상 값 10개를 포함
```

```python
loaded_reward_histories = []

# 저장된 파일 불러오기
index = 0
filename = f'080205_CartpoleResults/CQL_reward_history_{index}.pkl'
while os.path.exists(filename):

    with open(f'080205_CartpoleResults/CQL_reward_history_{index}.pkl', 'rb') as f:
        reward_history = pickle.load(f)
        loaded_reward_histories.append(reward_history)
    index += 1  # 파일이 이미 존재하면 인덱스 증가
    filename = f'080205_CartpoleResults/CQL_reward_history_{index}.pkl'  # 새로운 파일 이름 업데이트

Online_CQL = np.mean(loaded_reward_histories,axis=0)  # 각 리스트는 임의의 보상 값 10개를 포함
```

```python
# reward_history 시각화
plt.plot(Double_DQN, label = "Double DQN",linewidth=0.5)
plt.plot(DQN, label = "DQN",linewidth=0.5)
plt.plot(Online_CQL, label = "Online CQL",linewidth=0.5)
plt.title('Reward History')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.legend()
plt.show()
```
